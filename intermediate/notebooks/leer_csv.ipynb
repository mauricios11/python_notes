#### Pasos para cargar el dataset

**1-** importamos la librería csv y definimos una función que abra un archivo, teniendo la ruta (path) como parámetro.
* y en su interior colocamos el `with open(ruta, 'permiso') as archivo`

de la siguiente forma: 

        def read_csv(path):
             with open(path, 'r') as df:

**2-** dentro del `with` colocaremos una variable para leer el csv por medio de la función `reader()` de la librería *csv* teniendo como argumento:
* el nombre que indicamos en el *with*
* `delimiter = ','` aclarar la manera en la que van separados los datos (comas)

  así: `reader = csv.reader(df , delimiter = ',')`

  *la función reader( ) es un iterable*


**3-** indicar que se va a **leer fila por fila usando un ciclo** que itere sobre la variable que contiene el `reader( )`
    * colocamos un separador e imprimimos una fila por vez
  así: 
  
         for row in df:
            print('***' * 5)
            print(row)


```python
def read_csv(path):
    with open(path, 'r') as df: #abrir el csv
        reader = csv.reader(df, delimiter= ',') #leer el csv
        
        for row in reader:
            print('***' *5)
            print(row)
```

```python
read_csv(FILE)
```

Podemos observar que estamos leyendo cada una de las filas, cada una de ellas sería una lista
* ahora necesitamos convertir estas filas en un diccionario

#### Convertir las filas en un diccionario
**1-** Para eso hay que obtener los headers. Los cuales están en la primera fila del archivo
* Estando en la función que implementamos más arriba, podemos invocar a la primera fila usando un: `next(`*reader del csv*`)`

*estas serían las keys del diccionario*

**2-** definimos un ciclo que itere sobre el reader: *for fila in reader:*

**3-** unimos las variables del *header* y el *reader* por medio de la función `zip()`

  * Esto debería ser una lista, por lo que aplicamos un `list(zip(header, iterador del ciclo))`
    * colocamos esto en una nueva variable 
    
   así : 

        for fila in reader
          iterable = list(zip(header , fila))

**3-** por medio de un ciclo, iteramos sobre el *reader* e imprimimos la variable del *zip*

**4-** a partir de la variable que contiene sl *zip* vamos a hacer un dict comprehension, con los siguientes elementos:
* las keys serán los *headers*
* y los values será el iterador del ciclo (en este casi fila)
* estarán iterando sobre la variable que contiene el *zip*

  así: `diccionario_df = {header , fila for header , fila in iterable}`


**5-** después creamos una lista con los diccionarios que hicimos en el paso anterior
* para ello, en una nueva variable declaramos una lista. y dentro del *for* anterior indicamos que en cada iteración del diccionario, se agregue a la lista
así : `data = []` (dentro del for) `data.append(diccionario_df)`




```python
def read_header(path):
    with open(path, "r") as df: # abrimos el csv
        reader = csv.reader(df, delimiter= ",") # leemos el csv
        header = next(reader) # definimos los headers y los imprimimos
        print(f' Este es el header del dataset: \n')
        print(header)

read_header(FILE)     
```

```python
# hacemos un dict comprehension

def read_csv(path):
    with open(path, 'r') as df:
        reader = csv.reader(df, delimiter= ',')
        header = next(reader)
        country_df = []
        for row in reader:
            iterador = zip(header, row)
            #print(list(iterador))
            country_dict = {header : row for header, row in iterador}
            country_df.append(country_dict)
    return country_df
            
read_csv(FILE)
```
